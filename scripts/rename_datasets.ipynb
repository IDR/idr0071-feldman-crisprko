Datasets are prefixed with the experiment index to avoid duplicate dataset names across different experiments during the import. Run this to remove the prefix.

```python
import omero.clients
from omero.gateway import BlitzGateway
import os
import yaml
```

```python
with open(os.path.expanduser('~/blitzgateway-credentials.yml')) as f:
    creds = yaml.safe_load(f)
conn = BlitzGateway(**creds, secure=True)
c = conn.connect()
assert c
```

```python
projects = [p for p in conn.getObjects('Project') if p.name.startswith('idr0071')]
for p in projects:
    exp = p.name[-1]
    assert exp in 'ABCDEF'
    for d in p.listChildren():
        assert d.name.startswith(f'{exp} ')
        print(f'{p.name}: {d.name} → {d.name[2:]}')
        d.setName(d.name[2:])
        d.save()
```

```python
conn.close()
```
