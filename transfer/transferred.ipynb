# Create the file list for import

Some files are problematic, remove them from the main list and import them separately

```python
with open('failures.txt') as f:
    failures = set(f.read().split())
failure_import_lists = []
```

```python
basedir = '/uod/idr/filesets/idr0071-feldman-crisprko/20191104-gcloud/'
#experimentA/10X_c8-SBS-8/10X_c8-SBS-8_B2_Tile-503.sbs.tif

with open('idr0071-feldman-crisprko-files-unrenamed.txt') as f:
    local_filelist = set(f.read().split())

def rowfunc(r, exp):
    project = 'idr0071-feldman-crisprko/experiment{}'.format(exp)
    process = 'process' if (r['path'] == '{}/process'.format(exp)) else 'raw'
    cycleifset = (r['cycle'] + ' ') if r['cycle'] else ''
    # Prefix datasets with {exp} due to https://github.com/ome/omero-blitz/issues/91
    # They will have to be renamed later
    dataset = '{exp} {cycleifset}{tag} {well} {process}'.format(
        **r, exp=exp, cycleifset=cycleifset, process=process)
    filepath = r.file.replace('gs://idr_transfer/', basedir)

    return pd.Series([
        'Project:name:{}/Dataset:name:{}'.format(project, dataset),
        filepath,
        os.path.splitext(os.path.basename(r.file))[0].strip(),
    ])

for exp in 'ABCDEF':
    print('Experiment', exp)
    experiment = df.loc[(df['experiment'] == exp) & (df['ext'] == 'tif')]
    import_list = experiment.apply(partial(rowfunc, exp=exp), axis=1)
    for f in import_list.loc[:,1]:
        assert f in local_filelist, 'Not found: ' + f
    idx_failures = import_list.iloc[:,1].apply(lambda r: r in failures)
    failure_import_lists.append(import_list.loc[idx_failures==True, :])
    import_list.loc[idx_failures==False, :].to_csv('idr0071-experiment{}-filelist.tsv'.format(exp), sep='\t', header=False, index=False)
```

```python
failure_import_list = pd.concat(failure_import_lists)
failure_import_list[1] = failure_import_list[1].str.replace('20191104-gcloud', '20200623-fixes')
failure_import_list.to_csv(
    'idr0071-experimentRETRY-filelist.tsv'.format(exp), sep='\t', header=False, index=False)
```

## Repeat, but this time use tifffile to get the dimensions of each file

```python
from tifffile import imread

def rowfunc2(r, exp):
    project = 'idr0071-feldman-crisprko/experiment{}'.format(exp)
    process = 'process' if (r['path'] == '{}/process'.format(exp)) else 'raw'
    cycleifset = (r['cycle'] + ' ') if r['cycle'] else ''
    # Prefix datasets with {exp} due to https://github.com/ome/omero-blitz/issues/91
    # They will have to be renamed later
    dataset = '{exp} {cycleifset}{tag} {well} {process}'.format(
        **r, exp=exp, cycleifset=cycleifset, process=process)
    filepath = r.file.replace('gs://idr_transfer/', basedir)
    shape = imread(filepath, is_ome=False).shape

    return pd.Series([
        'Project:name:{}/Dataset:name:{}'.format(project, dataset),
        filepath,
        os.path.splitext(os.path.basename(r.file))[0].strip(),
        shape,
    ])

for exp in 'ABCDEF':
    print('Experiment', exp)
    experiment = df.loc[(df['experiment'] == exp) & (df['ext'] == 'tif')]
    import_list = experiment.apply(partial(rowfunc2, exp=exp), axis=1)
    for f in import_list.loc[:,1]:
        assert f in local_filelist, 'Not found: ' + f
    import_list.to_csv('idr0071-experiment{}-filelist-withshape.tsv'.format(exp), sep='\t', header=False, index=False)
```
